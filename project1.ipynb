Q1. What is the count of applications submitted by each country?

```python
data['country_citizenship'] =data['country_of_citizenship'].fillna('').astype(str)+data['country_of_citzenship'].fillna('').astype(str)
data.groupby(['country_citizenship']).size().sort_values(ascending=False).rename('No_of_application').reset_index()
```

Q2. List top 3 countries with highest number of applications and also show number of applications by them.

```python
data.groupby(['country_citizenship']).size().sort_values(ascending=False).head(3).rename('Applications').reset_index()
```

Q3. How many applications were denied?

```python
len(data[data['case_status']=='Denied'])
```

Q4. How many applicants holding H-1B visas were certified for Permenant Residency? (Consider all forms used for writing "H1-B" in data. Consider only H1-B, not H1-B1)

```python
len(data[(data.case_status=='Certified')&((data.class_of_admission=='H-1B')|(data.class_of_admission=='H1B')|(data.class_of_admission=='AOS/H-1B'))])
```

Q5. Sort columns in descending order according to number of missing values in them.

```python
data.isnull().sum().sort_values(ascending=False).rename('missing_val').reset_index()
```

Q6. List top 10 cities from which employers have submitted the most applications. How many applicarions were submitted by them?

```python
data.groupby(['employer_city']).size().sort_values(ascending=False).head(10).rename('no_of_application').reset_index()
```

Q7. Which are the top 10 employers who have submitted the highest number of applications? How many?

```python
data.groupby(['employer_name']).size().sort_values(ascending=False).head(10).rename('no_of_application').reset_index()
```

Q8. What is the average time of processing the applications?

```python
data['processing_time']=pd.to_datetime(data['decision_date'])-pd.to_datetime(data['case_received_date'])
data['processing_time'].dropna().mean()
```

Q9. What is count of applications for each case status(e.g. Certified, Denied etc) for applicants with "Doctorate" degree?

```python
data[data.foreign_worker_info_education=='Doctorate'].groupby('case_status').size().rename('count').reset_index()
```

Q.10 What are the top 3 economic sectors that has maximum applicants? How many applications are submitted from these sectors? (Use case_no column here, instead of case_number)

```python
data.groupby('us_economic_sector').size().rename('No_of_Applications').reset_index().sort_values('No_of_Applications',ascending=False).head(3)
```

Q.11 What is the mean wage of all the certified visa applications?

```python
data['wage']=data['wage_offer_from_9089'].fillna('0').astype(str)+data['wage_offered_from_9089'].fillna('0').astype(str)
data['unit']=data['wage_offer_unit_of_pay_9089'].fillna('').astype(str)+data['wage_offered_unit_of_pay_9089'].fillna('').astype(str)
data["wage"] = data["wage"].convert_objects(convert_numeric=True)
data[data['case_status']=='Certified'].wage.mean()
    
```
